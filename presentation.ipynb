```python
from Bio.Seq import Seq
import subprocess
from pathlib import Path
```

```python
path = Path('/beegfs/projects/ciwars/ILLUMINA_LANE_1/')
files = sorted([str(x) for x in path.glob('*') if x.is_file() and ".fastq" in x.suffixes])
if len(files) % 2 != 0:
    raise Exception("Miss pair-end files")

pairs = zip(files[::2], files[1::2])

pairs_1 = files[::2]
pairs_2 = files[1::2]
```

```python
subprocess.run(["megahit", "-1", ",".join(pairs_1), "-2", ",".join(pairs_2), "--presets", "meta-large", "-o", "."])

  
```
