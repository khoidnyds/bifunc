```python
from Bio.Seq import Seq
import subprocess
from pathlib import Path
import pandas as pd
from pyfaidx import Fasta
```

```python
database_path = "database/CARD/broadstreet/nucleotide_fasta_protein_homolog_model.fasta"
query_path = "results/Hospital_Wastewater/03-01--16-08-45/orfs.fa"
```

```python

subprocess.run(["bowtie2-build", database_path, "bowtie_db_build"])
subprocess.run(["bowtie2", "-x", "bowtie_db_build", "-f", "-U", query_path, "-S", "bowtie2_aligned.sam", "--end-to-end"])
subprocess.run(["samtools", "view", "-S", "-b", "bowtie2_aligned.sam", "-o", "bowtie2_aligned.bam"])
subprocess.run(["samtools", "sort", "-o" "bowtie2_aligned_sorted.bam", "bowtie2_aligned.bam"])
subprocess.run(["samtools", "coverage", "bowtie2_aligned_sorted.bam", "-o", "bowtie2_aligned.cov"])
subprocess.run(["samtools", "coverage", "-m", "bowtie2_aligned_sorted.bam", "-o", "bowtie2_aligned.his"])
```
